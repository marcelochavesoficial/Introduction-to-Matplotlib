```python
# Import packages:
import numpy as np
import matplotlib.pyplot as plt
```

**Knowing Matplotlib**

Many use, few understand. Matplotlib is the main visualization library for Python. Built on top of numpy arrays and designed to integrate with leading Data Science tools, Matplotlib was created in 2002 by John Hunter.

John was a neurobiologist who analyzed electrocorticography signals along with a team of researchers. As they used proprietary software and had only one license, the researcher created Matplotlib to meet this original need, inspired by the scripted interface that MATLAB provided.
When I said in the first line that many people use the library, but few actually understand it, I meant that they don't know the way the matplotlib architecture was designed.

**Matplotlib architecture**

Basically, the architecture of matplotlib is made up of 3 layers:
Scripting Layer
Artist Layer
Backend Layer
To understand how the package works, it is necessary to understand that Matplotlib's architecture was designed to allow its users to create, render and update Figure objects. These Figures are displayed on the screen and interact with events such as keyboard and mouse inputs. This type of interaction is performed at the backend layer.
Matplotlib allows you to create a graph composed of multiple different objects. It's as if it doesn't generate a single thing, but an image that is composed of several isolated pieces, such as the y-axis, y-axis, title, subtitles, among others. The ability to change all multiple objects is provided by layer artist. Look at the code below and see how we are dealing with these "multiple objects". We plot the data on the Cartesian plane, create a title, and label the x and y axes.

```python
# Generate demonstrative values:
np.random.seed(42)
x = np.arange(10)
y = np.random.normal(size=10)

# Plot the values:
plt.plot(x, y)
plt.title("Example")
plt.xlabel("Axle x")
plt.ylabel("Axle y")

plt.show()
```

**Object Oriented Approach**

According to the official Matplotlib documentation, the library plots its data in Figure-type objects (windows, Jupyter widgets, among others), and these objects can contain one or more Axes-type objects (an area where points can be specified in terms of x-y coordinates, polar coordinates, x-y-z, among others).
In this object-oriented approach, more customizations and chart types are possible.
The "object-oriented" style explicitly creates two objects, Figure and Axes, and calls methods on top of them.
The simplest way to create a figure with an axes is using pyplot.subplots, or plt.subplots for short.
Then, you can use the Axes.plot method to plot data on top of some axes.

```python
fig, ax = plt.subplots()
```

What was done above was to create a Figure that will contain all the plots (Axes). In this case, as nothing was specified, only 1 Figure and 1 Axes (plot) were created.
From there, manipulation and customization starts to be directly in the ax variable, calling its methods.

```python
x = np.arange(0, 100)

# Plot the values:
fig, ax = plt.subplots()
ax.plot(x, x, label="linear")
ax.plot(x, x**2, label="squared")
ax.set_title("Approach 00")
ax.set_ylabel("Values ​​in Y")
ax.legend()

# fig.show()
fig.show()
```

Ready! We have the same graph, but we gain much more manipulation control over it. See the example below, where I create 2 objects of type Axes to plot 2 different graphs in 1 single Figure.

```python
# Plot 2 graphs:
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(8,3))

# Graph 1:
ax[0].plot(x, x)
ax[0].set_title("Linear")
ax[0].set_xlabel("x")
ax[0].set_ylabel("y")

# Graph 2:
ax[1].plot(x, x**2)
ax[1].set_title("Squared")
ax[1].set_xlabel("x")
ax[1].set_ylabel("y")

fig.tight_layout();
```

**Pyplot Interface Approach**

If above we need to explicitly create figures and axes, there is another approach that delegates all responsibility for creating and managing them.
For this, you will use pyplot functions to plot. Here's how to use this pyplot-style.

```python
x = np.arange(10)

plt.plot(x, x)
```

If I want to add a title to the chart I can do it directly following the same approach:

```python
plt.plot(x, x)
plt.title("Linear")
plt.xlabel("Axle x")
plt.ylabel("Axle y")
```

Similarly, if I want to keep adding objects to Axes, I just add new functions sequentially:

```python
plt.plot(x, x, label="Linear")
plt.plot(x, x**2, label="Squared")
plt.title("Approach Pyplot")
plt.ylabel("Values ​​in y")
plt.xlabel("Values in x")
plt.legend()
plt.show()
```

This is because when I run any pyplot function such as plt.plot() or plt.title() one of the following things happens:

* Implicitly use current Figure and Axes (if they already exist)
* Creates new Figure and Axes objects (if they don't exist yet)

Brad Solomon highlights an excerpt from the official matplotlib documentation that reads as follows:


With pyplot, simple functions are used to add plot elements (lines, images, texts, and so on) to the current Axes, belonging to the current Figure.

**Which approach should you use?**

The official documentation considers both approaches to be equally powerful. You should feel free to adopt either one, but try not to mix approaches.

The documentation also suggests that the use of pyplot can be restricted to interactive plots (as is the case with the Jupyter notebook) and the Object Oriented approach as preferred for non-interactive plots (functions and scripts used in a larger project, for example).
